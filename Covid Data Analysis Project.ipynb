```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"], inplace=True, axis=1)
```

```python
covid_df['Date']= pd.to_datetime(covid_df['Date'],format='%Y-%m-%d')
```

```python
covid_df.head()
```

## To find the Active cases

```python
covid_df['Active_Cases']= covid_df['Confirmed']-(covid_df['Cured']+covid_df['Deaths'])
covid_df.tail()
```

## Using Pivot Table

```python
statewise= pd.pivot_table(covid_df, values=["Confirmed","Deaths","Cured"], index="State/UnionTerritory",aggfunc=max)
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise= statewise.sort_values(by="Confirmed", ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

## Top 10 active cases states

```python
top_10_active_states=covid_df.groupby(by="State/UnionTerritory").max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'], ascending=False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 states with most active case in india", size=20)
ax = sns.barplot(data=top_10_active_states.iloc[:10], y = 'Active_Cases', x = 'State/UnionTerritory', linewidth=2, edgecolor='red')
```

## Write Again in format

```python
top_10_active_states=covid_df.groupby(by="State/UnionTerritory").max()[['Active_Cases','Date']].sort_values(by=['Active_Cases'], ascending=False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 states with most active case in india", size=20)
ax = sns.barplot(data=top_10_active_states.iloc[:10], y = 'Active_Cases', x = 'State/UnionTerritory', linewidth=2, edgecolor='red')

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

## Top 10 States with high number of deaths

```python
top_10_States_Deaths = covid_df.groupby(by="State/UnionTerritory").max()[['Deaths','Date']].sort_values(by=['Deaths'], ascending=False).reset_index()
fig = plt.figure(figsize=(18,5))
plt.title("Top 10 states with most death case in india", size=20)
ax = sns.barplot(data=top_10_States_Deaths.iloc[:12], y = 'Deaths', x = 'State/UnionTerritory', linewidth=2, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Total number of Deaths")
plt.show()
```

```python
# remove rows where state = india
vaccine= vaccine_df[vaccine_df.State!='India']
```

```python
vaccine
```

```python
vaccine.rename(columns= {'Total Individuals Vaccinated':'Total'}, inplace= True)
vaccine.head()
```

# Top 5 Most Vaccinate States in India

```python
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac= max_vac.sort_values('Total', ascending = False)[:5]  #Top 5 states
max_vac
```

```python
fig = plt.figure(figsize=(12,6))
plt.title("Top 5 Vaccinated States in India", size=20)
x = sns.barplot(data=max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth=2, edgecolor='black')

plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```

# Least 5 Vaccinate States in India

```python
min_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
min_vac= min_vac.sort_values('Total', ascending = True)[:5]  #Top 5 states
min_vac
```

```python
fig = plt.figure(figsize=(12,6))
plt.title("Least 5 Vaccinated States in India", size=20)
x = sns.barplot(data=min_vac.iloc[:10], y = min_vac.Total, x = min_vac.index, linewidth=2, edgecolor='black')

plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
